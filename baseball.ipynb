```python
new_team_df = team_df.drop(columns=["div_id","ghome","wc_win","div_win","team_id_lahman45","team_id_retro","dp","hbp","sf","cs"])
new_team_df
```

```python
#league id 0

new_team_df['league_id'] = new_team_df['league_id'].fillna("None")
new_team_df['lg_win'] = new_team_df['lg_win'].fillna("NA")
new_team_df['ws_win'] = new_team_df['ws_win'].fillna("NA")
new_team_df['r'] = new_team_df['r'].fillna(0)
new_team_df['ab'] = new_team_df['ab'].fillna(0)
new_team_df['h'] = new_team_df['h'].fillna(0)
new_team_df['double'] = new_team_df['double'].fillna(0)
new_team_df['triple'] = new_team_df['triple'].fillna(0)
new_team_df['hr'] = new_team_df['hr'].fillna(0)
new_team_df['bb'] = new_team_df['bb'].fillna(0)
new_team_df['so'] = new_team_df['so'].fillna(0)
new_team_df['sb'] = new_team_df['sb'].fillna(0)
new_team_df['ra'] = new_team_df['ra'].fillna(0)
new_team_df['er'] = new_team_df['er'].fillna(0)
new_team_df['era'] = new_team_df['era'].fillna(0)
new_team_df['cg'] = new_team_df['cg'].fillna(0)
new_team_df['sho'] = new_team_df['er'].fillna(0)
new_team_df['sv'] = new_team_df['er'].fillna(0)
new_team_df['ipouts'] = new_team_df['er'].fillna(0)
new_team_df['ha'] = new_team_df['er'].fillna(0)
new_team_df['hra'] = new_team_df['er'].fillna(0)
new_team_df['bba'] = new_team_df['er'].fillna(0)
new_team_df['soa'] = new_team_df['er'].fillna(0)
new_team_df['park'] = new_team_df['park'].fillna("None")
new_team_df['attendance'] = new_team_df['attendance'].fillna(0)


new_team_df.head()
#new_team_df.column()
print(len(new_team_df.columns))

#new_team_df.loc[new_team_df["ws_win"] == "Y"]
```
