```python
import pandas as pd
from sqlalchemy import create_engine
```

### Data Cleaning

- Clean the allstar csv

```python
file='Resources/allstar.csv'

all_star_df=pd.read_csv(file)
```

```python
#show the dataframe
all_star_df=all_star_df.rename(columns={'year':'year_'})

all_star_df=all_star_df.drop(columns=["game_num"])
all_star_df.head()
```

```python
#replace na values with 0's for game played and started in the allstar game
all_star_df['gp']=all_star_df['gp'].fillna(0)
all_star_df['starting_pos']=all_star_df['starting_pos'].fillna(0)
all_star_df['game_id']=all_star_df["game_id"].fillna("NA")
```

```python
#show the new all_star dataframe
all_star_df
```

- Clean the batting csv

```python
file='Resources/batting.csv'

batting_df=pd.read_csv(file)
```

```python
#drop columns we don't need
batting_df=batting_df.drop(columns=['cs','hbp','sh','sf','g_idp','ibb'])
```

```python
#replace na values with 0's for game played and started in the allstar game
batting_df['league_id']=batting_df['league_id'].fillna(0)
batting_df["ab"]=batting_df["ab"].fillna(0)
batting_df["r"]=batting_df["r"].fillna(0)
batting_df["h"]=batting_df["h"].fillna(0)
batting_df["double_"]=batting_df["double_"].fillna(0)
batting_df["triple"]=batting_df["triple"].fillna(0)
batting_df["hr"]=batting_df["hr"].fillna(0)
batting_df["rbi"]=batting_df["rbi"].fillna(0)
batting_df["sb"]=batting_df["sb"].fillna(0)
batting_df["bb"]=batting_df["bb"].fillna(0)
batting_df["so"]=batting_df["so"].fillna(0)
```

```python
batting_df=batting_df.rename(columns={'year':'year_','g':'g_','double':'double_'})
batting_df
```

- Clean the batting csv

```python
file='Resources/pitching.csv'

pitching_df=pd.read_csv(file)
```

```python
pitching_df
```

```python
pitching_df=pitching_df.drop(columns=['ipouts','baopp','ibb','wp','bk','bfp','gf','sf','g_idp','sh'])
```

```python
pitching_df=pitching_df.rename(columns={'year':'year_','g':'g_'})
```

```python
#replace na values with 0's for game played and started in the allstar game
pitching_df['league_id']=pitching_df['league_id'].fillna("None")
pitching_df['w']=pitching_df['w'].fillna(0)
pitching_df['w']=pitching_df['l'].fillna(0)
pitching_df['g_']=pitching_df['g_'].fillna(0)
pitching_df['gs']=pitching_df['gs'].fillna(0)
pitching_df['cg']=pitching_df['cg'].fillna(0)
pitching_df['sho']=pitching_df['sho'].fillna(0)
pitching_df['sv']=pitching_df['sv'].fillna(0)
pitching_df['h']=pitching_df['h'].fillna(0)
pitching_df['er']=pitching_df['er'].fillna(0)
pitching_df['hr']=pitching_df['hr'].fillna(0)
pitching_df['bb']=pitching_df['bb'].fillna(0)
pitching_df['so']=pitching_df['so'].fillna(0)
pitching_df['era']=pitching_df['era'].fillna(0)
pitching_df['hbp']=pitching_df['hbp'].fillna(0)
```

```python
pitching_df
```

```python
batting_df.to_csv('clean_csv/clean_batting.csv')
```
