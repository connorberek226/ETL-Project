```python
import pandas as pd
```

### Data Cleaning

- Clean the allstar csv

```python
file='allstar.csv'

all_star_df=pd.read_csv(file)
```

```python
#show the dataframe
all_star_df=all_star_df.rename(columns={'year':'year_'})
all_star_df
```

```python
#replace na values with 0's for game played and started in the allstar game
all_star_df['gp']=all_star_df['gp'].fillna(0)
all_star_df['starting_pos']=all_star_df['starting_pos'].fillna(0)
```

```python
#show the new all_star dataframe
all_star_df
```

- Clean the batting csv

```python
file='batting.csv'

batting_df=pd.read_csv(file)
```

```python
#drop columns we don't need
batting_df=batting_df.drop(columns=['cs','hbp','sh','sf','g_idp','ibb','stint'])
```

```python
#replace na values with 0's for game played and started in the allstar game
batting_df['league_id']=batting_df['league_id'].fillna(0)
```

```python
batting_df=batting_df.rename(columns={'year':'year_','g':'g_','double':'double_'})
batting_df
```
